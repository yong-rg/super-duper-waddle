```python
import pandas as pd
import openpyxl
import duckdb
import re
```

```python
amz = 'https://docs.google.com/spreadsheets/d/1pP8jJDYPS1nHjHHDM1IycjtdFLLJIA8kRQ5AaC32Su0'
google = 'https://docs.google.com/spreadsheets/d/1oADSDDLjFmVs5tkc6KxJQW5B0VRg1qxaSqtQSV6EemA'


urls = [amz, google]
```

```python
duckdb.sql(f"""DROP TABLE IF EXISTS amz""")
duckdb.sql(f"""DROP TABLE IF EXISTS google""")
```

```python
print(amz)
u = f'{amz}/export?format=xlsx'

xl = pd.ExcelFile(u)

sheet_names = xl.sheet_names
num_sheets = len(sheet_names)
print(sheet_names)
print(num_sheets)

df = pd.read_excel(xl,2)

df = pd.DataFrame(df)

display(df)

duckdb.sql(f"""CREATE TABLE IF NOT EXISTS amz AS SELECT * FROM df""")
```

```python
print(google)
u = f'{google}/export?format=xlsx'

xl = pd.ExcelFile(u)

sheet_names = xl.sheet_names
num_sheets = len(sheet_names)
print(sheet_names)
print(num_sheets)

df = pd.read_excel(xl,3)

df = pd.DataFrame(df)

links = df.URL
asin_url = []

for l in links:
    asin = re.findall(r'(?<=\/)[A-Za-z0-9]{10}', l)
    asin_url.append(asin[0])

print(asin_url)

df['ASIN'] = asin_url

display(df)

duckdb.sql(f"""CREATE TABLE IF NOT EXISTS google AS SELECT * FROM df LIMIT 5""")
```

```python
# Create view 'google_agg' for Google data
duckdb.execute("""
                CREATE OR REPLACE VIEW google_agg AS
                SELECT DISTINCT
                    g.Campaign as campaign,
                    SUM(g.Cost) as cost,
                    SUM(g.Impressions) as impr,
                    SUM(g.Clicks) as clicks,
                    ROUND(AVG(g.CPC), 2) as cpc,
                FROM google as g
                GROUP BY g.campaign
                ORDER BY g.campaign ASC
""").df()
```

```python
# Create view 'amz_asin' for AMZ ASIN data
duckdb.execute("""
                SELECT DISTINCT
                    a.ASIN,
                    a.Campaign,
                    SUM(a.Sales)
                FROM amz as a
                GROUP BY a.ASIN, a.Campaign
                ORDER BY a.ASIN ASC
""").df()
```

```python
# Join Google campaign data + AMZ sales data
duckdb.execute("""
               SELECT
                g.campaign,
                g.cost,
                CAST(g.clicks as INT) clicks,
                g.cpc,
                CAST(SUM(a.DPV) as INT) as dpv,
                CAST(SUM(a.ATC) as INT) as atc,
                CAST(SUM(a.Purchases) as INT) as purchases,
                SUM(a.Sales) as sales
               FROM google_agg g
               LEFT JOIN amz a
               ON g.Campaign = a.Campaign
               GROUP BY g.campaign, g.cost, g.clicks, g.cpc,
               ORDER BY g.campaign ASC
""").df()
```
