```python
import pandas as pd
import openpyxl
import duckdb
import re
```

```python
amz = 'https://docs.google.com/spreadsheets/d/1pP8jJDYPS1nHjHHDM1IycjtdFLLJIA8kRQ5AaC32Su0'
google = 'https://docs.google.com/spreadsheets/d/1oADSDDLjFmVs5tkc6KxJQW5B0VRg1qxaSqtQSV6EemA'


urls = [amz, google]
```

```python
duckdb.sql(f"""DROP TABLE IF EXISTS amz""")
duckdb.sql(f"""DROP TABLE IF EXISTS google""")
```

```python
print(amz)
u = f'{amz}/export?format=xlsx'

xl = pd.ExcelFile(u)

sheet_names = xl.sheet_names
num_sheets = len(sheet_names)
print(sheet_names)
print(num_sheets)

df = pd.read_excel(xl,2)

df = pd.DataFrame(df)

display(df)

duckdb.sql(f"""CREATE TABLE IF NOT EXISTS amz AS SELECT * FROM df""")
```

```python
print(google)
u = f'{google}/export?format=xlsx'

xl = pd.ExcelFile(u)

sheet_names = xl.sheet_names
num_sheets = len(sheet_names)
print(sheet_names)
print(num_sheets)

df = pd.read_excel(xl,3)

df = pd.DataFrame(df)

links = df.URL
asin_url = []

for l in links:
    asin = re.findall(r'(?<=\/)[A-Za-z0-9]{10}', l)
    asin_url.append(asin[0])

print(asin_url)

df['ASIN'] = asin_url

display(df)

duckdb.sql(f"""CREATE TABLE IF NOT EXISTS google AS SELECT * FROM df""")
```

```python
duckdb.execute("""
                SELECT Campaign, SUM(COST) as Cost
                FROM google
                GROUP BY Campaign 
                ORDER BY Campaign
""").df()
```

```python
duckdb.execute("""
               SELECT Campaign, sum(Sales), sum(DPV), sum(ATC)
               FROM amz
               GROUP BY Campaign
""").df()
```

```python
# ASIN
duckdb.execute("""
                SELECT DISTINCT
                    g.ASIN,
                    g.Campaign,
                    g.Cost,
                    g.Impressions,
                    g.Clicks,
                    g.CPC,
                    a.DPV,
                    a.ATC,
                    a.Purchases,
                    a.Sales
                FROM google as g
                LEFT JOIN amz a
                ON g.ASIN = a.ASIN
                ORDER BY g.Campaign ASC
""").df()

# Ad Group

# Aggregate
```

```python
# Campaign
duckdb.execute("""
                SELECT DISTINCT
                    g.Campaign,
                    g.Cost as Cost,
                    g.Impressions,
                    g.Clicks,
                    g.CPC,
                    a.DPV,
                    a.ATC,
                    a.Purchases as Purchases,
                    a.Sales as Sales,
                    a.Sales / g.Cost as ROAS
                FROM google as g
                LEFT JOIN amz a
                ON g.Campaign = a.Campaign
""").df()
```
